## 2. Design a function named conversation_by_date(file, date) that will help us to extract the entire conversation which happened on a particular date. ## 6. Design a function named avg_length_by_person(file, person)that will help us to getthe average length of the messages posted by a particular person. As shown, this function shouldtake two arguments:a. file: A text file that contains the WhatsApp chat conversationb. person: A person of interest 

```python
#punctuation
import string
p = string.punctuation
p
```

```python
def avg_length_by_person(file,person):
    fhand =open(file,encoding='utf8')
    f = fhand.readlines()  #reading all lines
    count_p = 0   #taking variable to have the count lines with corresponding person
    line_start = []     #taking two lists to have the starting and ending point of the line having corresponding person 
    line_end = []
    chat_length=[] #taking a list to have the length of each chat done by corresponding person
    count_word=[0,0]  #taking a list to get the difference of cumulative counting of length 
    for line in f:
        line=line.rstrip()
        cond = (person in line[:line.find(":",line.find(":")+1)])
        #taking a condition to have the persons chat only
        if cond:
            temp = count_p+1           #taking a variable to count the end point   
            line_start.append(count_p) 
            for i in f[temp:]:
                if (i[5:8] == "/20"):
                    line_end.append(temp)
                    break
                temp +=1
            else:
                continue
        count_p+=1
    for start, end in zip(line_start, line_end): 
        #taking zip function to have the ideration of both start and end point in for loop
        # print(start,"\t",end)
        for line in f[start:end]:
            if line[5:8] == "/20":
                line=line[line.find(" ",line.find(":",line.find(":")+1)+1):].rstrip()
            # to extract only chat use find method to have the chat of lines having date after the space 
            # which belongs to the line next to the 2nd ":"
            else:
                line=line.rstrip()
            if "added"not in line:
                if "created"not in line: #removing the lines except chat
                    if "changed"not in line:
                        if "left"not in line: 
                                if "end-to-end encrypted"not in line:
                                    #remove the punctuation here
                                    t=str.maketrans("","",p)
                                    line=line.translate(t)                                
                                    words=line.split()   # spliting the lines into words
                                    for word in words:
                                        if word!=" ":  #removing the space
                                            count_word[0]+=1
                                    c_len=count_word[0]-count_word[1] #taking a variable to have the length of the lines        
                                    count_word[1]=count_word[0]      
                                    chat_length.append(c_len)
                                  # using append method to add the length of chat in the counting list  
    avg_length_by_person=int(sum(chat_length)/(len(chat_length)))
    # calculating the avg_length_by_person 
    return (avg_length_by_person) #printing avg_length_by_person
```
